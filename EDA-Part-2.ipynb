## Encoding Categorical Data
 ### One Hot Encoding
 


<font color=blue>Use to create dummy variables to replace the category in categorical variables

```python
df_dummies=pd.get_dummies(data1,prefix='Type1',columns=['Type'])
df_dummies.head()
```

```python
df_dummies.columns
```

### Sklearn Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

labelencoder=LabelEncoder()
df_dummies['Regionname']=labelencoder.fit_transform(df_dummies.Regionname)
```

```python
df_dummies['Regionname'].head(5)
```

```python
df_dummies['Regionname'].value_counts()
```

```python
df_dummies['Regionname'].unique()
```

```python
df_dummies['Regionname'].nunique()
```

```python
df_dummies.head(4)
```

### Sklearn OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

HOTencoder=OneHotEncoder()

encoded=HOTencoder.fit_transform(df_dummies.Regionname.values.reshape(-1,1)).toarray()

encoded
```

```python
encoded.shape
```

### Add this data frame back to df_dummies

```python
df_encoded=pd.DataFrame(encoded,columns=['Regionname_'+str(int(i)) for i in range(encoded.shape[1])])

df_encoded.head()
```

```python
df_encoded.shape
```

```python
df_dummies1= pd.concat([df_dummies,df_encoded],axis=1)

df_dummies1.head()
```

## Normalization and Scaling

##### StandardScaler normlized the data using the formula(x-mean)/standard  deviation

```python
from sklearn.preprocessing import StandardScaler

std_scale=StandardScaler()

std_scale
```

```python
data1['Rooms_std_scale']= std_scale.fit_transform(data1[['Rooms']])

data1['Rooms_std_scale'].head()
```

```python
data1['Rooms_std_scale'].min() ,data1['Rooms_std_scale'].max()
```

```python
data1['Rooms_std_scale'].describe()
```

### MinMaxScaler normalizes the data using formula (x-min)(max-min)

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scale=MinMaxScaler()
minmax_scale
```

```python
data1['Car_minmax']= minmax_scale.fit_transform(data1[['Car']]) 
```

```python
data1['Car_minmax'].head()
```

## Log Transform

```python
from sklearn.preprocessing import FunctionTransformer

log_trans=FunctionTransformer()
log_trans
```

```python
data1['Distance'].isnull().sum()
```

```python
data1['Distance']=data1['Distance'].fillna(data1['Distance'].mean())
data1['Distance'].head()
```

```python
data1['Distance_log']= log_trans.fit_transform(data1[['Distance']])

data1['Distance_log'].head()
```

```python
data1['Distance_log'].describe()
```

### Exponential Transformation

```python
exp_trans= FunctionTransformer(np.exp)
```

```python
exp_trans
```

```python
data1['Rooms_exp']=exp_trans.fit_transform(data1[['Rooms']])
```
